```python
%matplotlib inline 
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

# Filtrar imagens manualmente

```python
img = cv2.imread("./images/0/20170730_135420-(F01050)-0.jpeg")
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(imgGray,"gray") 
plt.show()
```

```python

limiteInferior = 100
limiteSuperior = 256

imgFiltrada = np.zeros( imgGray.shape )

for lin in range(len(imgGray)):
    for col in range(len(imgGray[lin])):
        if (imgGray[lin,col]>=limiteInferior and imgGray[lin,col]<=limiteSuperior):
            imgFiltrada[lin,col] = 1
           
plt.imshow(imgFiltrada) 
plt.show()

```

# Filtrando com OpenCV

```python
ret, imgFiltrada = cv2.threshold(imgGray, limiteInferior, limiteSuperior, cv2.THRESH_BINARY)

plt.imshow(imgFiltrada) 
plt.show()
```

# Jeito Python de ser

```python

imgFiltrada = np.copy(imgGray)

imgFiltrada[imgGray < limiteInferior ] = 0
imgFiltrada[imgGray > limiteSuperior] = 0
#imgFiltrada[imgGray > 0] = 1
           
plt.imshow(imgFiltrada) 
plt.show()

```
