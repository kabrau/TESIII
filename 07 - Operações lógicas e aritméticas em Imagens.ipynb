```python
#para mostrar imagens aqui no jupyter notebook
%matplotlib inline 

import numpy as np
import cv2
import matplotlib.pyplot as plt 
```

## Lendo Imagens

```python
img1 = cv2.cvtColor(cv2.imread("./images/lili.jpg"), cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(cv2.imread("./images/coracao.png"), cv2.COLOR_BGR2RGB)

plt.subplot(1,2,1)
plt.imshow(img1) 

plt.subplot(1,2,2)
plt.imshow(img2) 

plt.show()
```

# Adio de Imagens

```python
soma = cv2.add(img1,img2)
somaNP = img1+img2

print("img 1:\n", img1[150:151,200:203])
print("img 2:\n", img2[150:151,200:203])
print("opencv Add:\n", soma[150:151,200:203])
print("numpy Sum:\n", somaNP[150:151,200:203] ) 

```

```python
plt.subplot(1,2,1)
plt.imshow(soma) 

plt.subplot(1,2,2)
plt.imshow(somaNP) 

plt.show()
```

# Subtrao

```python

subtracao = cv2.subtract(img1, img2)
subtracao2 = cv2.subtract(img2, img1)

plt.subplot(1,2,1)
plt.imshow(subtracao) 

plt.subplot(1,2,2)
plt.imshow(subtracao2) 

plt.show()
```

# Image Blending (Misturado)

```python
imgBlend1 = cv2.addWeighted(img1, 0.4, img2, 0.6, 0)
imgBlend2 = cv2.addWeighted(img1, 0.8, img2, 0.2, 0)

plt.subplot(1,2,1)
plt.imshow(imgBlend1) 

plt.subplot(1,2,2)
plt.imshow(imgBlend2) 

plt.show()
```

# Novas Imagens 


```python
imgCoracao = np.copy(img2)
imgCoracao[imgCoracao > 1] = 255

plt.subplot(1,2,1)
plt.imshow(imgCoracao) 

imgRetangulo = np.copy(img2)
imgRetangulo[:] = 0
imgRetangulo[150:250, 150:350] = 255

plt.subplot(1,2,2)
plt.imshow(imgRetangulo) 

plt.show()
```

# AND, OR, XOR, NOT

```python

plt.subplot(2,3,1)
plt.title("AND")
plt.imshow(cv2.bitwise_and(imgCoracao,imgRetangulo)) 

plt.subplot(2,3,2)
plt.title("OR")
plt.imshow(cv2.bitwise_or(imgCoracao,imgRetangulo)) 

plt.subplot(2,3,3)
plt.title("XOR")
plt.imshow(cv2.bitwise_xor(imgCoracao, imgRetangulo)) 

plt.subplot(2,3,4)
plt.title("NOT Coracao")
plt.imshow(cv2.bitwise_not(imgCoracao)) 

plt.subplot(2,3,5)
plt.title("NOT Retangulo")
plt.imshow(cv2.bitwise_not(imgRetangulo)) 


plt.subplot(2,3,6)
plt.title("(NOT) AND")
plt.imshow(cv2.bitwise_and(cv2.bitwise_not(imgCoracao),imgRetangulo)) 

plt.subplots_adjust(hspace=.5)

plt.show()
```
