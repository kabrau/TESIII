```python
%matplotlib inline 

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

ler dataset


```python
path_0 = "./images/0/"
path_1 = "./images/1/"

listX_0 = []
listY_0 = []

for filename in os.listdir(path_0):
    img = cv2.imread(os.path.join(path_0, filename))
    cv_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    listX_0.append(cv_rgb)
    listY_0.append(0)

listX_1 = []
listY_1 = []

for filename in os.listdir(path_1):
    img = cv2.imread(os.path.join(path_1, filename), cv2.COLOR_BGR2RGB)
    cv_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    listX_1.append(cv_rgb)
    listY_1.append(1)
    
print("Total Images classe 0 ", len(listX_0))    
print("Total Images classe 1 ", len(listX_1))    
    
print("Shape da primeira imagem classe 0 ",listX_0[0].shape)    
print("Shape da primeira imagem classe 1 ",listX_1[1].shape)    
```

Mostrar Imagens

```python
plt.subplot(1,2,1) # total lines, total cols, position
plt.imshow(listX_0[0])    
plt.subplot(1,2,2) # total lines, total cols, position
plt.imshow(listX_1[0])    
plt.show()   
```

Ver histograma

```python
color = ('r','g','b')

#plt.subplot(1,2,1) # total lines, total cols, position
for i,col in enumerate(color):
    histr = cv2.calcHist([listX_0[0]],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])

plt.show()

#plt.subplot(1,2,2) # total lines, total cols, position
for i,col in enumerate(color):
    histr = cv2.calcHist([listX_1[0]],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
    

plt.show()
```
