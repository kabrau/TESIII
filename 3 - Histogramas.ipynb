```python
%matplotlib inline 
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

# Calcular Histograma manualmente

```python
img = cv2.imread("./images/0/20170730_135420-(F01050)-0.jpeg")
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(imgGray.shape)

histogram = np.zeros( 256 )
print(histogram.shape)

pixels = imgGray.ravel() #retorna um array 1d continuo
#print(pixels.shape)
#pixels2 = imgGray.reshape(3519) #modifica o shape da matriz
#print(pixels2.shape)
#print(pixels[0:10])
#print(pixels2[0:10])

for px in pixels:
    histogram[px] += 1
```

```python
plt.subplot(2,1,1) # total lines, total cols, position
plt.imshow(imgGray,"gray") 

plt.subplot(2,1,2) # total lines, total cols, position
plt.plot(histogram)
plt.xlim([0,256])
plt.show()
```

# Calculando histogramas com funcoes

```python
histogram2 = cv2.calcHist([imgGray],[0],None,[256],[0,256])

plt.subplot(2,1,1) # total lines, total cols, position
plt.xlim([0,256])
plt.plot(histogram2) 
plt.show()
```

# Mostrando histogram de imagem RGB


```python
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(imgRGB) 
plt.show()

color = ('r','g','b')

for i,col in enumerate(color):
    histr = cv2.calcHist([imgRGB],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])

plt.show()
```
