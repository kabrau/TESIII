# Exercício de reconhecimento de dígitos

```python
import matplotlib.pyplot as plt
from sklearn.externals import joblib
from sklearn import datasets
from skimage.feature import hog
from sklearn.svm import LinearSVC
import numpy as np
```

# Carrega o dataset de dígitos

* São 70.000 imagens de dígitos (0 a 9) escritos a mão
* Cada dígito tem o tamanho de 28x28, mas é armazenado em uma array de 784 bytes (28x28)


```python
dataset = datasets.fetch_mldata("MNIST Original")

features = np.array(dataset.data, 'int16') # 70.000 Digitos 
labels = np.array(dataset.target, 'int')   # 70.000 Rótulos
```

# Mostrando alguns dígitos

```python
plt.subplot(1,4,1)
plt.imshow(features[0].reshape((28, 28)), cmap=plt.cm.gray)
plt.subplot(1,4,2)
plt.imshow(features[10000].reshape((28, 28)), cmap=plt.cm.gray)
plt.subplot(1,4,3)
plt.imshow(features[20000].reshape((28, 28)), cmap=plt.cm.gray)
plt.subplot(1,4,4)
plt.imshow(features[50000].reshape((28, 28)), cmap=plt.cm.gray)
plt.show()
```

```python
list_hog_fd = []
for feature in features:
    fd = hog(feature.reshape((28, 28)), block_norm='L2', orientations=9, pixels_per_cell=(14, 14), cells_per_block=(1, 1), visualise=False)
    list_hog_fd.append(fd)
hog_features = np.array(list_hog_fd, 'float64')
```

```python
clf = LinearSVC()
```

```python
clf.fit(hog_features, labels)
```

```python
joblib.dump(clf, "digits_cls.pkl", compress=3)
```

```python
# Import the modules
import cv2
from sklearn.externals import joblib
from skimage.feature import hog
import numpy as np

# Load the classifier
clf = joblib.load("digits_cls.pkl")

# Read the input image 
im = cv2.imread("./images/photo_2.jpg")

# Convert to grayscale and apply Gaussian filtering
im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY) 

plt.imshow(im_gray, cmap=plt.cm.gray)
plt.show()

im_gray = cv2.GaussianBlur(im_gray, (5, 5), 0)

plt.imshow(im_gray, cmap=plt.cm.gray)
plt.show()

# Threshold the image
ret, im_th = cv2.threshold(im_gray, 90, 255, cv2.THRESH_BINARY_INV)

plt.imshow(im_th, cmap=plt.cm.gray)
plt.show()

```

```python
# Find contours in the image
#ctrs, hier = cv2.findContours(im_th.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
im2, ctrs, hier = cv2.findContours(im_th.copy(),cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
# Get rectangles contains each contour
rects = [cv2.boundingRect(ctr) for ctr in ctrs]
```

```python
# Read the input image 
im = cv2.imread("./images/photo_2.jpg")

# For each rectangular region, calculate HOG features and predict
# the digit using Linear SVM.
for rect in rects:
    # Draw the rectangles
    cv2.rectangle(im, (rect[0], rect[1]), (rect[0] + rect[2], rect[1] + rect[3]), (0, 255, 0), 3) 

cv2.imshow("Resulting Image with Rectangular ROIs", im)
cv2.waitKey()
```

```python


# For each rectangular region, calculate HOG features and predict
# the digit using Linear SVM.
for rect in rects:
    # Draw the rectangles
    cv2.rectangle(im, (rect[0], rect[1]), (rect[0] + rect[2], rect[1] + rect[3]), (0, 255, 0), 3) 
    # Make the rectangular region around the digit
    leng = int(rect[3] * 1.6)
    pt1 = int(rect[1] + rect[3] // 2 - leng // 2)
    pt2 = int(rect[0] + rect[2] // 2 - leng // 2)
    roi = im_th[pt1:pt1+leng, pt2:pt2+leng]
    # Resize the image
    roi = cv2.resize(roi, (28, 28), interpolation=cv2.INTER_AREA)
    roi = cv2.dilate(roi, (3, 3))
    # Calculate the HOG features
    roi_hog_fd = hog(roi, block_norm='L2', orientations=9, pixels_per_cell=(14, 14), cells_per_block=(1, 1), visualise=False)
    nbr = clf.predict(np.array([roi_hog_fd], 'float64'))
    cv2.putText(im, str(int(nbr[0])), (rect[0], rect[1]),cv2.FONT_HERSHEY_DUPLEX, 2, (0, 255, 255), 3)

cv2.imshow("Resulting Image with Rectangular ROIs", im)
cv2.waitKey()
```
