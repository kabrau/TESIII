# Bibliotecas

```python
%matplotlib inline 
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
```

# Espaços e Cores

```python
img = cv2.imread("./images/1/8169723-0.jpg")

#Converter BGR para RGB
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

print(img[0,0])
print(imgRGB[0,0])
```

```python
plt.subplot(1,2,1) # total lines, total cols, position
plt.imshow(img) 
plt.subplot(1,2,2) # total lines, total cols, position
plt.imshow(imgRGB) 
plt.show()
```

# Converter BGR para GrayScale


* Convertendo manualmente 


```python
print(imgRGB.shape)

#imgGray = np.zeros((69,51), int)
imgGray = np.zeros(imgRGB.shape[:-1])

print(imgGray.shape)

for lin in range(imgRGB.shape[0]):
    for col in range(imgRGB.shape[1]):
        #gray = int((imgRGB[lin,col,0] + imgRGB[lin,col,1] + imgRGB[lin,col,2])/3)
        gray = int(imgRGB[lin,col,0]*0.2989  + imgRGB[lin,col,1]*0.5870 + imgRGB[lin,col,2]*0.1140)
        imgGray[lin,col] = gray

#print(imgGray)        
    
```

```python
plt.imshow(imgGray, "gray") 
plt.show()        
        
```

* Convertendo usando a função

```python
imgGray2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(imgGray2.shape)

print(imgGray[0,0:10])
print(imgGray2[0,0:10])

plt.imshow(imgGray2, "gray") 
plt.show() 
```

# Outros Espaços de Cor

```python
imgYCrCb = cv2.cvtColor(img, cv2.COLOR_BGR2YCR_CB)
imgHSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
imgLAB = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
imgLUV = cv2.cvtColor(img, cv2.COLOR_BGR2LUV)
```

```python
print("BGR", img[0,0])
print("RGB", imgRGB[0,0])
print("GrayScale", imgGray[0,0])
print("YCrCb", imgYCrCb[0,0])
print("HSV", imgHSV[0,0])
print("LAB", imgLAB[0,0])
print("LUV", imgLUV[0,0])
```

## Voltando para RGB

```python
imgYCrCb2RGB = cv2.cvtColor(imgYCrCb, cv2.COLOR_YCR_CB2RGB)

print("RGB original", imgRGB[0,0])
print("YCrCb", imgYCrCb[0,0])
print("RGB retornado", imgYCrCb2RGB[0,0])
```
