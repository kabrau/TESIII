# Criar um classificador simples para Sinaleiras 
* O dataset está em:
    * ./images/0    __=> Sinaleiras abertas__
    * ./images/1    __=> Sinaleiras fechadas__
    
    
A ideia é que se houver um determinado __número/percentual de pixels verdes__ na imagem, será considerada __ABERTA__
E se houver um determinado __número/percentual de pixels vermelhos__ na imagem, será considerada __FECHADA__


```python
#bibliotecas
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

1) Ler imagens e tipo em duas listas
na __listaX__ carregar todas as imagens das duas pastas
na __listaY__ é um array de inteiros, com a mesma quatidade de elementos do listaX, informando as classes 0 e 1 (aberto, fechado)


```python
listaX = []
listaY = []

path_0 = "./images/0/"
path_1 = "./images/1/"

for filename in os.listdir(path_0):
    img = cv2.imread(os.path.join(path_0, filename))
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    listaX.append(imgRGB)
    listaY.append(0)
    
for filename in os.listdir(path_1):
    img = cv2.imread(os.path.join(path_1, filename))
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    listaX.append(imgRGB)
    listaY.append(1)
    

```

2) Separar metade para testes, pulando de 2 em 2


```python
# converte para numpy array
listaY = np.array(listaY)
listaX = np.array(listaX)
    
listaX_Treino = listaX[::2]
listaY_Treino = listaY[::2]

listaX_Teste = listaX[1::2]
listaY_Teste = listaY[1::2]

print("Total Images: ", len(listaY) )
print("Classe 0 (Aberto):", (listaY == 0).sum() )
print("Classe 1 (Fechado):", (listaY == 1).sum() )

print("Total Treino: ", len(listaY_Treino), " Teste: ", len(listaY_Teste))



```

3) Ideia 1 - mostrando o histograma das classes




```python
def calculaHistograma(imagem):
    countR = cv2.calcHist([imagem],[0],None,[256],[0,256])
    countG = cv2.calcHist([imagem],[1],None,[256],[0,256])
    countB = cv2.calcHist([imagem],[2],None,[256],[0,256])
    return countR, countG, countB
    
    
hitogram_0_R = np.zeros( (256,1) )
hitogram_0_G = np.zeros( (256,1) )
hitogram_0_B = np.zeros( (256,1) )

hitogram_1_R = np.zeros( (256,1) )
hitogram_1_G = np.zeros( (256,1) )
hitogram_1_B = np.zeros( (256,1) )


for indice, imagem in enumerate(listaX_Treino):
    countR, countG, countB = calculaHistograma(imagem)
    
    if (listaY_Treino[indice]==0):
        hitogram_0_R += countR
        hitogram_0_G += countG
        hitogram_0_B += countB
    else:
        hitogram_1_R += countR
        hitogram_1_G += countG
        hitogram_1_B += countB
    
#print(hitogram_0_R.ravel())
#print(hitogram_1_R.ravel())
```

```python

plt.title("R")

plt.plot(hitogram_0_R,color = "G")
plt.xlim([0,256])

plt.plot(hitogram_1_R,color = "R")
plt.xlim([0,256])

plt.show()


plt.title("G")

plt.plot(hitogram_0_G,color = "G")
plt.xlim([0,256])

plt.plot(hitogram_1_G,color = "R")
plt.xlim([0,256])

plt.show()



plt.title("B")

plt.plot(hitogram_0_B,color = "G")
plt.xlim([0,256])

plt.plot(hitogram_1_B,color = "R")
plt.xlim([0,256])

plt.show()
```

# Metodo mais simples do mundo

Olhando os histogramas, dá para ver que:
* para a classe O (aberto), há um pico do canal G, para pixels acima de 250
* para a classe 1 (fechado), há um pico do canal R, para pixels acima de 250

then 
 se a quantidade de pixels de valor __acima de 250__, no __canal G__, for maior que __10% do total__ de pixel, é ABERTO (classe 0)


```python
def classificaImagem(imagem):
    countR, countG, countB = calculaHistograma(imagem)
    totalPixels = imagem.shape[0]*imagem.shape[1]
    #print(totalPixels)
    
    limiteInferior = 250
    
    qtdG = np.sum(countG[limiteInferior:])
    qtdR = np.sum(countR[limiteInferior:])
    #print(qtd)
    
    percentual = 0.05
    
    classeRS = 1
    if (qtdG>0):
        if (totalPixels / qtdG > percentual):
            classeRS = 0
        
    #print(classeRS)
    return classeRS
    
```

## percorre todas imagens

```python
resultado = []

for indice, imagem in enumerate(listaX_Treino):
    classe = classificaImagem(imagem)
    resultado.append(classe)
    
    
```

## avalia

```python
def avaliar(listaY_certo, listaY_predita):
    
    desempenho = np.sum(listaY_certo == listaY_predita) #Soma/Conta o que é igual

    return desempenho/listaY_certo.shape[0]
```

```python
resultado = np.array(resultado)

print( avaliar(listaY_Treino, resultado) )
```

```python

resultado = []

for indice, imagem in enumerate(listaX_Teste):
    classe = classificaImagem(imagem)
    resultado.append(classe)
    
resultado = np.array(resultado)

print( avaliar(listaY_Teste, resultado) )
```
