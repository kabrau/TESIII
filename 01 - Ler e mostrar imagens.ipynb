# Mostrar várias imagens juntas

```python
img1 = cv2.imread("./images/0/20170701_105311-0.jpg")
img2 = cv2.imread("./images/0/20170730_135420-(F01050)-0.jpeg")
img3 = cv2.imread("./images/1/8169723-0.jpg")
img4 = cv2.imread("./images/1/20170701_172243-0.jpg")

print(img1.shape)
print(img2.shape)
print(img3.shape)
print(img4.shape)

imgRGB1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
imgRGB2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
imgRGB3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
imgRGB4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)

plt.subplot(2,3,1) # total lines, total cols, position
plt.imshow(imgRGB1) 

plt.subplot(2,3,2) # total lines, total cols, position
plt.imshow(imgRGB2) 

plt.subplot(2,3,3) # total lines, total cols, position
plt.imshow(imgRGB3) 

plt.subplot(2,3,4) # total lines, total cols, position
plt.imshow(imgRGB4) 

plt.show()
```

# Ler imagens de uma pasta

```python
path_0 = "./images/0/"

lista = []
for filename in os.listdir(path_0):
    img = cv2.imread(os.path.join(path_0, filename))
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    lista.append(imgRGB)

    
print("Total de imagens lidas:", len(lista))
```

## Ler apenas imagens


```python
path_0 = "./images/0/"

valid_images = [".jpg",".gif",".png",".tga"] #,".jpeg"]

lista = []
for filename in os.listdir(path_0):
    
    name, ext = os.path.splitext(filename)
    if ext.lower() not in valid_images:
        continue    
    
    img = cv2.imread(os.path.join(path_0, filename))
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    lista.append(imgRGB)

    
print("Total de imagens lidas:", len(lista))
```

```python
for i in range(0,10):
    plt.subplot(3,4,i+1) # total lines, total cols, position
    plt.imshow(lista[i]) 

plt.show()    
```
