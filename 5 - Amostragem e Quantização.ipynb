```python
%matplotlib inline 
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread("./images/0/20170730_135420-(F01050)-0.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

# Amostragem 
## Reduzindo a imagem

```python
# Seleciona uma em cada 2 colunas, e de cada coluna uma a cada duas linhas
n = 2
img_red = img[::n,::n]
print("Original", img.shape)
print("Reduzida", img_red.shape)

plt.title('original')
plt.imshow(img) 
plt.show()

plt.title('reduzida')
plt.imshow(img_red) 
plt.show()
```

## Aumentando a imagem

 

```python
# Os pixels da imagem atual sero duplucados no eixo x e y. Assim, a imagem volta a ter o tamanho original, 
# mas a partir da imagem reduzida

# Funo: np.repeat(matriz, vezes, eixo). O eixo 0  a altura e 1 a largura.
m = 2
img_aum = np.repeat(img_red, m, axis=0)
img_aum = np.repeat(img_aum, m, axis=1)

print("Reduzida", img_red.shape)
print("Aumentada", img_aum.shape)

plt.title('original')
plt.imshow(img) 
plt.show()

plt.title('reduzida')
plt.imshow(img_red) 
plt.show()
 
plt.title('aumentada')
plt.imshow(img_aum) 
plt.show()
    
```

# Quantizao

```python
# 255 / 31 = 8,22...
# Assim, teremos uma imagem com 8 tons de cinza. 
# A conta  feita desta forma para descartar a parte decimal dos nmeros e alterar o vetor para que possua apenas 8 valores possveis.

r = 31
img_qtd = np.uint8(img / r) * r
 
print("parte original", img.ravel()[100:120])
print("parte quantificada", img_qtd.ravel()[100:120])

    
plt.title('original')
plt.imshow(img) 
plt.show()

plt.title('quantificada')
plt.imshow(img_qtd) 
plt.show()    
    
```
